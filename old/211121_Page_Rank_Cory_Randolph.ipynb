# Page Rank

## Formula

General assumptions for this example of the Page Rank algorithm are that we are not using a damping factor and that we will only do one iteration to get the overall rank of the pages rather than multiple interactions which would narrow down to more precise ranking scores.

General Page Rank Fomula:


\begin{equation}
        PR(u) = \sum_{v \in B_u} 
        \frac{PR(v)}{L(v)}
\end{equation}

Note: $PR()$ is the Page Rank and $L()$ is the number of outbound links

## Calculation

Page rank formula for our specific problem:

\begin{equation}
        PR(A) = \frac{PR(B)}{L(B)} +
        \frac{PR(C)}{L(C)} + 
        \frac{PR(D)}{L(D)} +
        \frac{PR(E)}{L(E)}
\end{equation}



For the first iteration we assume that each node has an equal page rank of $\frac{1}{5} = 0.2$. Also, assume that a node that has no outbound nodes connecting to the nod in question has a Page Rank of 0.

\begin{equation}
        PR(A) = \frac{PR(B)}{L(B)} +
        \frac{PR(C)}{L(C)} + 
        \frac{PR(D)}{L(D)} +
        \frac{PR(E)}{L(E)} \\[1em]
        = \frac{\frac{1}{5}}{1} +
        0 + 
        \frac{\frac{1}{5}}{4} +
        \frac{\frac{1}{5}}{1} \\[1em]
        = \frac{9}{20} \\[1em]
        = 0.45
  \end{equation}
