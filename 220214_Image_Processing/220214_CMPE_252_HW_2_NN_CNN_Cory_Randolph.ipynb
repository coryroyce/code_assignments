We use below filters/kernels to do edge detection. Please determine what kind of edge  each one can detect. 

Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import signal
from scipy import ndimage
```

Load and Process the image

```python
def rgb2gray(rgb_image):
    return np.dot(rgb_image[...,:3], [0.299, 0.587, 0.114])

# Load the image
# image = mpimg.imread("/content/Iguana_Image.jpeg")
image = mpimg.imread("/content/San_Francisco_Image.jpeg")
image_gr = rgb2gray( image )

figureSize = (12,10)
plt.figure("Original Image", figsize=figureSize)
plt.imshow(image_gr)
plt.set_cmap("gray")
```

```python
filter_vertical = np.array([[ -1 , 0 , 1 ] ,
                      [ -1 , 0 , 1 ] ,
                      [ -1 , 0 , 1 ] ])

filter_horizontal = np.array([[1, 1, 1],
                      [ 0,  0,  0],
                      [ -1,  -1,  -1]])
       

# Apply filters
image_gr_norm = image_gr / 255.
image_filter_vertical = signal.convolve2d( image_gr_norm , filter_vertical ,'same')
image_filter_horizontal = signal.convolve2d( image_gr_norm , filter_horizontal ,'same')

# Display Image
plt.figure("Original", figsize=figureSize)
plt.title("Orginal Image")
plt.imshow(image_gr_norm)

plt.figure("Vertically Filtered Image", figsize=figureSize)
plt.title("Vertically Filtered Image")
plt.imshow(image_filter_vertical)

plt.figure("Horizontally Filtered Image", figsize=figureSize)
plt.title("Horizontally Filtered Image")
plt.imshow(image_filter_horizontal)
```
